```python
# it is clear that 1st entry (city of london) and  everything after 33 entry are not a bourughs. we can remove them
df.iloc[:,0]
```

```python
drop_index_list = [1]+ list(range(34,len(df)+1,1))  # 0 is the index of city of london and the rest are after 33rd index 
mod_df = df.drop(drop_index_list) # dropping all the non bourough entries and storing it into new data frame

mod_df.reset_index(drop=True)
```

```python
# melting the data frame 
mod_df = mod_df.melt(id_vars=['borough','postal_code'],var_name='date',value_name='avg_price')
```

```python
mod_df.info()
```

---
* From the above information on the columns it is pretty clear that 
    * we need to convert the avg_price to Floating type
    *  we can convert postal_code and borough into categorical type since there are only 32 unique type of variations and it will optimize the data frame
---

```python
mod_df['avg_price'] = mod_df.avg_price.astype(float)

# we can convert borough and postal_code to category columns too to optimize the data frame
mod_df[['borough','postal_code']] = mod_df[['borough','postal_code']].apply(lambda x: x.astype('category'))
```

```python
mod_df.info()
```

```python
# checking to see if there are any null values
mod_df.isnull().sum()
```

```python
# populating month column
mod_df['month'] = mod_df['date'].dt.month_name()
mod_df['year'] = mod_df['date'].dt.year
```

```python
mod_df.head()
```

---
>Making a new dataframe with statistical information about percentage change and annual percentage change
---

```python
# method to calculate the percentage increase in the price 
def perc_increase(series):
    
    min_val = series.iloc[0]
    max_val = series.iloc[-1]
    return ((max_val-min_val)/min_val)*100

# method to calculate the annual percentage increase in the price
def avg_annual_inc(series):
    _max_val = series.iloc[-1]
    _min_val = series.iloc[0]
    _yr_diff = series.index[-1].year - series.index[0].year
    
    return ((_max_val-_min_val)/_min_val)*(1/_yr_diff)*100
```

```python
stats_df = pd.DataFrame(mod_df.set_index('date').sort_index().groupby('borough')['avg_price'].agg([perc_increase, avg_annual_inc]))   # Populating stats data frame with percentage increase and annual perrcentage increase
stats_df.rename(columns={'perc_increase':'% increase', 'avg_annual_inc': '% annual increase'},inplace=True)  # Renaming the columns
```

```python
# Adding the latest current average price 
stats_df['latest_curr_avg_price'] = mod_df.set_index('date').groupby('borough')['avg_price'].agg(lambda x: x.loc[max(x.index)])
```

```python
ylim = (stats_df.latest_curr_avg_price.min(), stats_df.latest_curr_avg_price.max())
yticks = [i for i in range(int(stats_df.latest_curr_avg_price.min()-10000), int(stats_df.latest_curr_avg_price.max()+80000), int(stats_df.latest_curr_avg_price.min()/2))]
ylabels = list(map(lambda x: '{}M'.format(x/10**6) if x> 10**6 else '{}K'.format(x/(10**3)), yticks))
ax = stats_df.plot(kind='bar',y='latest_curr_avg_price',figsize=(20,10), ylim=ylim)
plt.yticks(yticks, ylabels)
plt.title('Latest Average housing price')
plt.ylabel('Average price (Euros)')
plt.show()
```
